```python
excel_file = "C:/Users/91703/Desktop/project/AD/ad1 and layers90.xlsx"  # Provide the path to your Excel file
df = pd.read_excel(excel_file, header=None)
data_array = df.values
# Plot the data
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Assuming your data has three columns: X, Y, and Z
ax.scatter(df.iloc[:, 0], df.iloc[:, 1], df.iloc[:, 2])

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
tic = time.time()
res = ripser(data_array, n_perm=600, maxdim=2)
toc = time.time()
timefull = toc-tic
print("Elapsed Time Subsampled Point Cloud: %.3g seconds"%(toc-tic))
dgms_sub = res['dgms']
idx_perm = res['idx_perm']
r_cover = res['r_cover']
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.scatter(data_array [:, 0], data_array [:, 1])
plt.title("Original Point Cloud (%i Points)"%(data_array.shape[0]))
plt.axis("equal")
plt.subplot(122)
plt.scatter(data_array [idx_perm, 0],data_array [idx_perm, 1])
plt.title("Subsampled Cloud (%i Points)"%(idx_perm.size))
plt.axis("equal")
plt.show()
```
